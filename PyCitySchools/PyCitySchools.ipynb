```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
sch_data_loaded = "Resources/sch_cplt.csv"
stu_data_loaded = "Resources/stu_cplt.csv"

# Read School and Student Data File and store into Pandas DataFrames
sch_data = pd.read_csv(sch_data_loaded)
stu_data = pd.read_csv(stu_data_loaded)

# Combine the data into a single dataset.  
sch_cplt_df = pd.merge(stu_data, sch_data, how="left", on=["school_name", "school_name"])
sch_cplt_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the number of schools in the DataFrame
school_count = len(sch_cplt_df["school_name"].unique())

# Calculate the number of students in the DataFrame
student_count = len(sch_cplt_df["student_name"])

# Calculate the total budget in the DataFrame
budget_count = sum(sch_cplt_df.budget.unique())

# Calculate the average math score in the DataFrame
avg_math_score = (sch_cplt_df["math_score"].mean())

# Calculate the average reading score in the DataFrame
avg_reading_score = (sch_cplt_df["reading_score"].mean())

# Calculate the percentage of students with a passing math score (70 or greater) in the DataFrame
num_passing_math = sch_cplt_df.loc[sch_cplt_df['math_score'] > 69]['math_score'].count()
percentage_math_pass = num_passing_math/student_count*100

# Calculate the percentage of students with a passing reading score (70 or greater) in the DataFrame
num_passing_reading = sch_cplt_df.loc[sch_cplt_df['reading_score'] > 69]['reading_score'].count()
percentage_reading_pass = num_passing_reading/student_count*100

# Calculate the percentage of students who passed math and reading (% Overall Passing) in the DataFrame
percentage_pass_overall = sch_cplt_df[(sch_cplt_df['math_score'] > 69) & (sch_cplt_df['reading_score'] > 69)]['student_name'].count()/student_count*100

# Place all of the data found into a summary DataFrame
district_summary_df = pd.DataFrame({"Total Schools": [school_count],
                              "Total Students": [student_count],
                              "Total Budget": [budget_count],
                              "Average Math Score": [avg_math_score],
                              "Average Reading Score":[avg_reading_score],
                              "% Passing Math":[percentage_math_pass],
                              "% Passing Reading":[percentage_reading_pass],
                              "% Overall Passing":[percentage_pass_overall]})

# Cleaner data format
district_summary_df['Total Students'] = district_summary_df.apply(lambda x: "{:,.0f}".format(x['Total Students']), axis=1)
district_summary_df['Total Budget'] = district_summary_df.apply(lambda x: "${:,.2f}".format(x['Total Budget']), axis=1)
district_summary_df['Average Math Score'] = district_summary_df.apply(lambda x: "{:.2f}%".format(x['Average Math Score']), axis=1)
district_summary_df['Average Reading Score'] = district_summary_df.apply(lambda x: "{:,.2f}%".format(x['Average Reading Score']), axis=1)
district_summary_df['% Passing Math'] = district_summary_df.apply(lambda x: "{:,.2f}".format(x['% Passing Math']), axis=1)
district_summary_df['% Passing Reading'] = district_summary_df.apply(lambda x: "{:,.2f}".format(x['% Passing Reading']), axis=1)
district_summary_df['% Overall Passing'] = district_summary_df.apply(lambda x: "{:,.2f}".format(x['% Overall Passing']), axis=1)

# Create and Display Dataframe
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Group by School Name
school_list = sch_cplt_df.set_index("school_name").groupby(["school_name"])

# List the different school types
type_of_sch = sch_data.set_index('school_name')['type']
#type_of_sch

# List the total number of students at each school
stu_numbers = school_list['Student ID'].count()
#stu_numbers

# Calculate total school budget
tot_sch_budget = sch_data.set_index('school_name')['budget']
#tot_sch_budget

# Calculate budget per student
budget_per_stu = sch_data.set_index('school_name')['budget']/sch_data.set_index('school_name')['size']
#budget_per_stu

# Calculate average math scores by school
sch_avg_math = school_list['math_score'].mean()
#sch_avg_math

# Calculate average reading scores by school
sch_avg_reading = school_list['reading_score'].mean()
#sch_avg_reading

# Calculate the percentage of students with a passing math score (70 or greater) in each school
sch_pass_math = sch_cplt_df[sch_cplt_df['math_score'] > 69].groupby('school_name')['Student ID'].count()/stu_numbers*100 
#sch_pass_math

# Calculate the percentage of students with a passing reading score (70 or greater) in each school
sch_pass_read = sch_cplt_df[sch_cplt_df['reading_score'] > 69].groupby('school_name')['Student ID'].count()/stu_numbers*100 
#sch_pass_read

# Calculate the percentage of students with a passing math and reading score (70 or greater) in each school
sch_overall = sch_cplt_df[(sch_cplt_df['math_score'] > 69) & (sch_cplt_df['reading_score'] > 69)].groupby('school_name')['Student ID'].count()/stu_numbers*100
sch_overall

# Place all of the data found into a summary DataFrame
school_summary_df = pd.DataFrame({"School Type": type_of_sch,
                            "Total Students": stu_numbers,
                            "Per Student Budget": tot_sch_budget,
                            "Total School Budget": budget_per_stu,
                            "Average Math Score": sch_avg_math,
                            "Average Reading Score": sch_avg_reading,
                            "% Passing Math": sch_pass_math,
                            "% Passing Reading": sch_pass_read,
                            "Overall Passing Rate": sch_overall})

# Cleaner data format
school_summary_df['Total Students'] = school_summary_df.apply(lambda x: "{:,.0f}".format(x['Total Students']), axis=1)
school_summary_df['Per Student Budget'] = school_summary_df.apply(lambda x: "${:,.2f}".format(x['Per Student Budget']), axis=1)
school_summary_df['Total School Budget'] = school_summary_df.apply(lambda x: "${:,.2f}".format(x['Total School Budget']), axis=1)
school_summary_df['Average Math Score'] = school_summary_df.apply(lambda x: "{:.2f}%".format(x['Average Math Score']), axis=1)
school_summary_df['Average Reading Score'] = school_summary_df.apply(lambda x: "{:,.2f}%".format(x['Average Reading Score']), axis=1)
school_summary_df['% Passing Math'] = school_summary_df.apply(lambda x: "{:,.2f}".format(x['% Passing Math']), axis=1)
school_summary_df['% Passing Reading'] = school_summary_df.apply(lambda x: "{:,.2f}".format(x['% Passing Reading']), axis=1)
school_summary_df['Overall Passing Rate'] = school_summary_df.apply(lambda x: "{:,.2f}%".format(x['Overall Passing Rate']), axis=1)

# Create and Display Dataframe
school_summary_df
```
